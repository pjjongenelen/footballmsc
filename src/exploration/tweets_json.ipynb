### Exploratory analysis of json files with tweets for each fixture
Currently very basic, but no worrying signals show

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ROOT = "C:/Users/timjo/python_projects/footballmsc"
```

# RKC - Sparta
Let's start with the smallest tweet set. Based on file size this is RKC Waalwijk - Sparta.

```python
rkcspa = pd.read_json(f"{ROOT}/data/tweets_json/rkcspa.json", lines=True)
print(f"Number of tweets in smallest dataframe: {rkcspa.shape[0]}")
rkcspa.head()
```

```python
pd.Series([d.get('username') for d in rkcspa.user]).value_counts()
```

So we have around 70 people tweeting about this match with more than 100 tweets - we should probably exclude 'RKCWAALWIJK' and 'SpartaRotterdam'.  
Those are number that we should be able to work with.  
Let's check the tweet volume through time:

```python
fig = plt.figure(figsize=(15,5))
rkcspa.groupby([rkcspa['date'].dt.day, rkcspa['date'].dt.hour]).size().plot(kind='bar')
plt.ylabel('# tweets')
plt.xlabel('Time of day (hour, on 22 and 23 oct 2021)')
plt.xticks(np.arange(20), [8, 10, 12, 15, 17, 5, 6, 7, 8, 11, 13, 15, 16, 17, 18, 19, 20, 21, 22, 23], rotation=0)
plt.axvline(14.5, color = 'green')
plt.axvline(16.5, color = 'green')
plt.show()
```
